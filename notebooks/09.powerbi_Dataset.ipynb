```python
import pandas as pd
```

```python
# Main dataset
df_main = pd.read_csv(r"C:\Users\Hey!\OneDrive\Desktop\Customer churn prediction & retention Platform\Data\processed_data\telecom_customer_churn_cleaned.csv")

# Risk segmentation
df_risk = pd.read_csv(r"C:\Users\Hey!\OneDrive\Desktop\Customer churn prediction & retention Platform\Risk_segmentation\outputs\customer_risk_segments.csv")

# Retention recommendations
df_retention = pd.read_csv(r"C:\Users\Hey!\OneDrive\Desktop\Customer churn prediction & retention Platform\retention_Intelligence\outputs\customer_retention_recommendations.csv")
```

```python
print(df_main.shape)
print(df_risk.shape)
print(df_retention.shape)
```

```python
df_main["Churn_Probability"] = df_risk["Churn_Probability"]

df_main["Risk_Segment"] = df_risk["Risk_Segment"]
```

```python
df_main["Retention_Recommendation"] = (df_retention["Retention_Recommendation"])

df_main["Potential_Revenue_Saved"] = (df_retention["Potential_Revenue_Saved"])
```

```python
df_main["Predicted_Churn"] = (df_main["Churn_Probability"] >= 0.50).astype(int)
```

```python
df_main.to_csv("C:\\Users\\Hey!\\OneDrive\\Desktop\\Customer churn prediction & retention Platform\\Power_BI_dashboard\\Dataset\\dashboard_master_dataset.csv",index=False)

print("Dashboard Dataset Created Successfully")
```
